```python
# Inheritance
# Inheritance â€” a key concept in OOP that lets one class reuse the features of another.
# Inheritance allows a new class (child) to inherit the attributes and methods of an existing class (parent).

# Example: We have a base class Car, and we want to create a more specific class ElectricCar.


# Step 1: Base Class â€” Car

class Car:
    def __init__(self, brand, color, model_year, mileage):
        self.brand = brand
        self.color = color
        self.model_year = model_year
        self.mileage = mileage

    def start(self):
        print(f"{self.color} {self.brand} ({self.model_year}) is starting...")

    def stop(self):
        print(f"{self.color} {self.brand} has stopped.")

    def drive(self, speed):
        print(f"{self.color} {self.brand} is driving at {speed} km/h.")

    def horn(self):
        print("Beep beep!")

    def show_info(self):
        print(f"Brand: {self.brand}")
        print(f"Color: {self.color}")
        print(f"Model Year: {self.model_year}")
        print(f"Mileage: {self.mileage} km/l")


car_1= Car("Maruti", "White", 2023, 18)


# Step 2: Child Class â€” ElectricCar Inheriting from Car

class ElectricCar(Car):
    def __init__(self, brand, color, model_year, mileage, battery_range):
        super().__init__(brand, color, model_year, mileage)
        self.battery_range = battery_range

    def charge(self):
        print(f"{self.brand} is charging... Battery range: {self.battery_range} km")

    # Optionally override a method
    def horn(self):
        print("Electric Beep!")

# Here:
# ElectricCar gets all features from Car
# super().__init__() calls the constructor of the parent class
# charge() is a new method
# We overrode horn() to change its behavior


# Step 3: Use Both Classes

my_ev = ElectricCar("Tesla", "Black", 2024, 0, 450)

my_ev.start()
my_ev.horn()
my_ev.drive(100)
my_ev.charge()
my_ev.stop()
my_ev.show_info()


car_1.start()
car_1.horn()
car_1.show_info()
```

```python
# Core OOP Concepts
    # Class: A blueprint for creating objects. It defines attributes (data) and methods (functions).
    
    # Object: An instance of a class with specific values.
    
    # Encapsulation: Hides internal details and exposes only necessary parts through methods.
    
    # Inheritance: Allows one class to use the properties and methods of another.
    
    # Polymorphism: Ability to redefine methods in derived classes.
```

```python
# ðŸ’» Example: OOP with All Key Concepts



# âœ… CLASS and ENCAPSULATION
class Person:
    def __init__(self, name, age):
        self.name = name              # public attribute
        self.__age = age              # private attribute (encapsulation)

    def show_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.__age}")   # accessing private variable inside the class

    def get_age(self):               # getter method
        return self.__age

    def set_age(self, new_age):      # setter method
        if new_age > 0:
            self.__age = new_age
        else:
            print("Invalid age!")

# âœ… INHERITANCE
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)   # call constructor of parent class
        self.student_id = student_id

    # âœ… POLYMORPHISM (overriding the show_info method)
    def show_info(self):
        print("=== Student Info ===")
        print(f"ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Age: {self.get_age()}")  # using getter from parent class

# âœ… OBJECT CREATION & TESTING
person1 = Person("Alice", 30)
student1 = Student("Bob", 20, "ST102")

# Show person info
print("=== Person Info ===")
person1.show_info()

# Show student info (polymorphism)
student1.show_info()

# Try accessing private variable directly (should fail)
# print(person1.__age)  # âŒ This will raise an error

# Use getter and setter to access private data safely
print("\nUsing getter:", person1.get_age())
person1.set_age(35)
print("Updated age:", person1.get_age())
```

```python
# Practice


class trees:
    def __init__(self, name, years):
        self.name = name
        self.__years = years  # private attribute

    def get_years(self):
        return self.__years

    def set_years(self, new_years):
        if new_years > 0:
            self.__years = new_years
        else:
            print('Invalid')

    def show(self):
        print(f'name: {self.name}')
        print(f'years: {self.__years}')


# Creating a tree object
first_tree = trees('palm', 56)
print(first_tree.get_years())       # Accessing years safely
first_tree.set_years(60)            # Updating safely
first_tree.show()                   # Display info


# Inherited class with height
class Short(trees):
    def __init__(self, name, years, height):
        super().__init__(name, years)
        self.height = height

    def show(self):
        print(f'name: {self.name}')
        print(f'years: {self.get_years()}')  # Safe access to private variable
        print(f'height: {self.height}')


# Create an object from Short class
tree_1 = Short('banyan', 78, 890)
tree_1.show()
first_tree.show()


# Create a list of tree objects
tree_list = [
    trees("Mango", 45),
    trees("Neem", 30),
    Short("Guava", 25, 300),
    Short("Jackfruit", 50, 500),
    trees("Coconut", 40)
]

# Loop through the list and display info
print("\n=== All Trees in the List ===")
for t in tree_list:
    t.show()
    print("------------------\n")
```

```python
# ðŸ“š Mini Project: Library Book Management System

# Goal: Create a simple Python program using OOP that lets you:
    #     Add books to the library
    #     Borrow a book
    #     Return a book
    #     Display available books

class Books:
    def __init__(self,title,author):
        self.title=title
        self.author=author
        self.available=True             #This sets the initial status of the book as available (not borrowed).
                                        # True means it can be borrowed. If a user borrows the book, weâ€™ll change this to False.


    def display(self):
        status ='Available' if self.available else 'Borrowed'       # This is a short if-else (called a ternary operator).
                                                                    # If self.available == True, then status = 'Available'
                                                                    # If self.available == False, then status = 'Borrowed'
        print(f'{self.title} by {self.author}-{status}')

b1 = Books("Wings of Fire", "A.P.J. Abdul Kalam")
b1.display()


class Library:                         # Defines a new class named Library.
                                       # This class will manage books â€” like adding, displaying, and keeping track of availability.
    def __init__(self):
        self.books = []                # Initializes the library with an empty list of books.


    def add_book(self, title, author):          # This method lets you add a new book to the library.
                                                # It takes two inputs: title and author.
        book = Books(title, author)              # It creates a new book object using the Books class (must be defined separately).
                                                # The new book has the given title and author.
        self.books.append(book)                 # Adds the new book object to the self.books list â€” i.e., it's now part of the library.
       
        print(f"\n Book '{title}' added to the library.")

    def show_books(self):                       # Defines a method to show all books that are available (not borrowed).
        print("\nðŸ“š Available Books:")          # Prints a title for the output.                                               
        found = False                           # This variable is used to track whether any available book was found.
                                                # Initially, we assume none are found (False).
        for book in self.books:                 # Loops through each book in the library's list.
            if book.available:
                book.display()                  # If it is available, show its info using the display() method from the Book class.
                found = True                    # Set found to True since we found at least one available book.
        if not found:
            print("No books are currently available.")

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower() and book.available:    
                        # Title match: Convert both titles to lowercase using .lower() to make the search case-insensitive.
                        # Available: Only borrow the book if itâ€™s marked as available (True).
                book.available = False   # Mark the book as borrowed by setting available to False.
                print(f"\nâœ… You borrowed '{book.title}'.")
                return
                # Print confirmation and use return to exit the method early (since the book is already found and borrowed).
        print("\nâŒ Sorry, that book is not available.")

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower() and not book.available:
                book.available = True
                print(f"\nâœ… You returned '{book.title}'.")
                return
        print("\nâŒ That book wasn't borrowed or doesn't exist.")


# --------- MENU SYSTEM -------------
def main():
    library = Library()            # Creates an object library from the Library class.       
                                   # This object will be used to call methods like add_book, show_books, etc.

    while True:                    # Starts an infinite loop to keep showing the menu until the user chooses to exit.
        print("\n=== ðŸ“– Library Menu ===")
        print("1. Add Book")
        print("2. Show Available Books")
        print("3. Borrow Book")
        print("4. Return Book")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            title = input("Enter book title: ")
            author = input("Enter author name: ")
            library.add_book(title, author)

        elif choice == "2":
            library.show_books()

        elif choice == "3":
            title = input("Enter the title of the book to borrow: ")
            library.borrow_book(title)

        elif choice == "4":
            title = input("Enter the title of the book to return: ")
            library.return_book(title)

        elif choice == "5":
            print("Exiting the library system. Goodbye! ðŸ‘‹")
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

# Run the menu
main()
```

```python
import pandas as pd
from datetime import datetime

df= pd.read_csv('OMNI2014.csv')
df
```

```python
# Define the OOP Class

class SpaceWeatherObservation:
    def __init__(self, year, doy, hour, minute, speed, density, bz, symh):
        self.year = int(year) #Take the year value passed during object creation. Store it in the object's self.year attribute so it can be used later.
        self.doy = int(doy)
        self.hour = int(hour)
        self.minute = int(minute)
        self.speed = float(speed)
        self.density = float(density)
        self.bz = float(bz)
        self.symh = int(symh) 

    def datetime(self):
        return datetime.strptime(f"{self.year} {self.doy} {self.hour} {self.minute}", "%Y %j %H %M")

#  This method converts the year, DOY (day-of-year), hour, and minute into a datetime object using datetime.strptime().
# Why this matters: It helps for plotting time-series data or comparing dates easily.
        # %Y = 4-digit year
        # %j = day of year (001 to 365/366)
        # %H = hour (00â€“23)
        # %M = minute (00â€“59)
    
    def is_storm(self):
        return self.symh < -50                #âž¡ Returns True if SYM/H index is less than -50.


    def is_anomalous_speed(self):
        return self.speed > 500               #âž¡ Returns True if speed is higher than 500 km/s 


    def is_negative_bz(self):
        return self.bz < -5                   #âž¡ Returns True if IMF Bz is strongly negative

    def summary(self):                        #âž¡ This prints out the complete status of the observation in a readable format.
        dt = self.datetime()                  #âž¡ Converts the DOY, hour, minute into a datetime object for display.
        flags = []                            #âž¡ Initializes an empty list to store flags like â€œStormâ€, â€œHigh Speedâ€, etc.
        if self.is_storm():
            flags.append("âš ï¸ Storm")
        if self.is_anomalous_speed():
            flags.append("High Speed")
        if self.is_negative_bz():
            flags.append("Bz < -5")            #âž¡ Each of these checks conditions using earlier methods and adds the relevant label if true.
        status = " | ".join(flags) if flags else "Normal"   #âž¡ Joins all flags together into a string. If no flags, status is â€œNormalâ€.
        print(f"[{dt}] Speed: {self.speed} km/s | Density: {self.density} cm^-3 | Bz: {self.bz} nT | SYM/H: {self.symh} â†’ {status}")


#[{dt}]: dt is the output of the method self.datetime(), which converts the date (Year + DOY + Hour + Minute),
               # into a proper Python datetime object like: [2014-04-10 12:30:00]
# â†’ {status}: status is a summary of what was detected.
# It may show:
#             Normal
#             âš ï¸ Storm
#             High Speed
#             Bz < -5
#             or any combination, separated by |.

# print(...) function inside your class method didnâ€™t print anything while the code ran. 
    # The line is inside the .summary() method.
    # If you just created the objects and didnâ€™t run .summary() for each, nothing will print.


```
