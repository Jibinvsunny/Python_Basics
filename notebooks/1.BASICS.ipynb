```python
# Break and Return

# break: Exits a loop only
    # It stops the loop immediately and jumps outside the loop.
    # Commonly used in for and while loops.

for i in range(1, 10):
    if i == 5:
        break
    print(i)



# return: Exits a function
    # It ends the function completely and can return a value.
    # If you call the function, it sends the result back to the caller.

def greet(name):
    if name == "":
        return "No name entered"
    return f"Hello, {name}!"

print(greet("Jithin"))
print(greet(""))

# return exits the function immediately, no more lines after it will run.



def find_even(numbers):
    for n in numbers:
        if n % 2 == 0:
            print("Even number found!")
            return n  # exits the function
        else:
            print("Not even, checking next...")

print(find_even([1, 3, 7, 4, 5]))
```

```python
# Combined Example:

def find_even_with_break(lst):
    for x in lst:
        if x % 2 == 0:
            print(f"Even found: {x}")
            break
    print("Loop ended (but function still running)")

def find_even_with_return(lst):
    for x in lst:
        if x % 2 == 0:
            return f"Even found: {x}"
    return "No even number"

print(find_even_with_break([1,3,5,4,5]))
print(find_even_with_return([1,3,5,4,5]))
print(find_even_with_return([1,3,7,5]))
```

```python
# Create a function that returns the first number in a list that is divisible by 7.

def find_divisible_by_7(numbers):
    for n in numbers:
        if n % 7 == 0:
            return n
    return "No number divisible by 7"

# Test cases
print(find_divisible_by_7([10, 13, 21, 25]))  # âžœ 21
print(find_divisible_by_7([5, 11, 19]))       # âžœ No number divisible by 7


# Same task using break instead of return

def find_divisible_by_7_with_break(numbers):
    for n in numbers:
        if n % 7 == 0:
            print(f"First number divisible by 7: {n}")
            break
    else:
        print("No number divisible by 7")

find_divisible_by_7_with_break([10, 13, 21, 25])  # âžœ prints 21
find_divisible_by_7_with_break([5, 11, 19])       # âžœ prints "No number divisible by 7"
```

```python
# found = False (or True) flag is very common in programming when you need to track whether a certain condition has been metâ€”especially in loops.

numbers = [2, 4, 6, 8, 10]
search = 6

found = False  # Assume not found yet

for n in numbers:
    if n == search:
        found = True
        break  # Exit the loop once found

if found:
    print("âœ… Number found!")
else:
    print("âŒ Number not found.")
```

```python
# enumerate() is a built-in Python function used when looping through an iterable and you want both:
#     The index (position) of each item
#     The value of the item itself

# Basic Syntax: enumerate(iterable, start=0)
# iterable: Any iterable like a list, tuple, string, etc.
# start (optional): The index you want to start counting from (default is 0)


fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits):
    print(index, fruit)

print('\n---------with start-----------\n')

for index, fruit in enumerate(fruits, 1):
    print(index, fruit)
```

```python
# Lambda Functions

# Lambda functions are small, anonymous functions defined using the lambda keyword.
# Best used when the function is short-lived and doesn't need a name.

# Single argument
square = lambda x: x**2
print(square(5))  


# Multiple arguments
add = lambda x, y: x + y
print(add(3, 4))  


squares =lambda x,y: (x**2 ,y**2)
print(squares(5,6))


# With map
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)


# With filter
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)


# With sorted
points = [(1, 2), (3, 1), (5, 0)]
sorted_points = sorted(points, key=lambda x: x[1])
print(sorted_points)
```

```python
# List Comprehensions

# List comprehensions are a Pythonic way to create a new list by applying an expression to each item in an iterable (like a list or range). 
# Itâ€™s often more concise and faster than using a traditional for loop.


# Squares of numbers from 0 to 9
squares = [x**2 for x in range(10)]
print(squares)


# Convert temperatures from Celsius to Fahrenheit
celsius = [0, 10, 20, 34.5]
fahrenheit = [(x * 9/5) + 32 for x in celsius]
print(fahrenheit)


# Filter even numbers from a list
numbers = [1, 2, 3, 4, 5, 6]
evens = [n for n in numbers if n % 2 == 0]
print(evens)


# You can use a conditional expression in the comprehension itself:
labels = ['Even' if x % 2 == 0 else 'Odd' for x in range(5)]
print(labels)



# Suppose you want only speeds above 500 km/s
import pandas as pd
df= pd.DataFrame({
    'DOY': [1, 2, 3, 4, 5],
    'Speed': [400, 520, 680, 550, 600]
    })
high_speed = [s for s in df['Speed'] if s > 500]
print(high_speed)

```

```python
#It is a collection which is unordered, unindexed and unchangeable.

s = {'animal','bat','flower',5,'mother'}
print(s)
print(type(s))
```

```python
#Items cannot be changed, but can be added or removed.
#Duplicates are not allowed.
#set can contain int, float, string and boolen.
#set cannot contain list and dict because they are mutable.
#set can contain tuple (imutable).
#nesting is not possible- set cannot contain a set
```

```python
#set constructor
s = set(('apple','mango','orange'))
print(s)
print(type(s))
```

```python
#empty set
x = {}
print(x)
print(type(x))

y = set()
print(y)
print(type(y))
```

```python
#Access items
s = {'animal','bat','flower',5,'mother'}
for x in s:
    print(x) #unordered
```

```python
#Check if an item is present in set
s = {'animal','bat','flower',5,'mother'}
print('animal' in s)
print('girl' in s)
```

```python
#add and delete items from set
s = {'animal','bat','flower',5,'mother'}
s.add('boy') #add a specific item
print(s)


l = {78,89,'beauty'}
s.update(l) #merge two sets
print(s)


s.remove('boy') #remove a specific item
print(s)

s.pop() #remove a random item
print(s)

s.clear() #empty a set
print(s)
```

```python
s = {'animal','bat','flower',5,'mother'}
l = {78,89,'beauty'}
x =s.union(l)
print(x)
```

```python
# return a new set of identical items from two sets - INTERSECTION
x = {5,6,7,8,9,56}
y = {67,90,7,6,8,88}
print(x.intersection(y))
```

```python
# Error Handling

# Error handling is the process of responding to runtime errors (also called exceptions) in your program without crashing it. 
# This is especially important when dealing with:

            # User input
            # File handling
            # Network operations
            # Data analysis


# try:
    # Code that may raise an exception
# except ExceptionType:
    # Code to handle the exception
# finally:
    # Code that runs no matter what


try:
    number = int(input("Enter a number: "))
    result = 10 / number
    print(result)
except ZeroDivisionError:           #This block only runs if the code inside try: causes a division by zero.
                                    #Example: If user enters 0, this message is printed.
    print("You cannot divide by zero!")
except ValueError:                  #This block runs if the user enters text or symbols that can't be converted to an integer.

    print("Invalid input. Please enter a number.")
finally:                            #finally: Always runs, no matter what happened in try or except.

    print("Execution finished")
```

```python
# datetime

# The datetime module in Python is a built-in module used for working with dates and times-
     # â€” like today's date, time now, date math, formatting, and more.

# Use datetime:
#     Get the current date and time
#     Add or subtract days/weeks
#     Format dates (e.g., "20-Jun-2025")
#     Calculate age, duration, deadlines, etc.


import datetime


now = datetime.datetime.now()     # Get Current Date and Time
print("Current date and time:", now) 


today = datetime.date.today()     # Get Just the Date
print("Todayâ€™s date:", today)


d = datetime.date(2023, 12, 25)     # Create a Custom Date
print("Christmas:", d)
```

```python
# Date Difference (using timedelta)

from datetime import date, timedelta

d1 = date(2025, 6, 20)
d2 = date(2025, 7, 1)

diff = d2 - d1
print("Days difference:", diff.days)


# Add Days to a Date

future_date = datetime.date.today() + datetime.timedelta(days=10)
print("10 days from now:", future_date)
```

```python
# Format a Date (using strftime)

now = datetime.datetime.now()
print(now.strftime("Today is %d-%b-%Y"))


# Format Code	Meaning               Example
#    %d	        Day (2-digit)	      20
#    %m	        Month (2-digit)       06
#    %b	        Month name short	  Jun
#    %Y	        Year (4-digit)	      2025
#    %H	        Hour (24-hour)	      14
#    %I	        Hour (12-hour)	      02
#    %p	        AM/PM	              PM

```

```python
# Parse Date from String (using strptime)

dt = datetime.datetime.strptime("25-12-2023", "%d-%m-%Y")
print("Parsed date:", dt.date())
```

```python
# Code to Get DOY

from datetime import date

# Input year, month, day
year = 2025
month = 6
day = 20

# Create a date object
given_date = date(year, month, day)

# Get the DOY using .timetuple().tm_yday
doy = given_date.timetuple().tm_yday

print(f"Date: {given_date} â†’ Day of Year: {doy}")
```

```python

# Get Date from Year and DOY

from datetime import date, timedelta

def get_date_from_doy(year, doy):
    start = date(year, 1, 1)
    return start + timedelta(days=doy - 1)
print(get_date_from_doy(2025, 171))

```
