```python
# Filtering in Pandas

import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']
})

# Filter rows where Age > 30
filtered_df = df[df['Age'] > 30]
print(filtered_df)

print(type(filtered_df))
```

```python
# Grouping in Pandas

import pandas as pd


data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales'],
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [50000, 55000, 60000, 62000, 70000]
}
df = pd.DataFrame(data)

# Group by department and calculate mean salary
grouped = df.groupby('Department')['Salary'].mean()
print(grouped)
```

```python
# Group by multiple columns

df.groupby(['Department', 'Employee'])['Salary'].sum()
```

```python
# Using multiple aggregation functions

df.groupby('Department')['Salary'].agg(['mean', 'max', 'min'])
```

```python
# Filtering groups

# Example: keep departments where avg salary > 60,000


x = df.groupby('Department').filter(lambda x: x['Salary'].mean() > 60000)


```

```python
# Iterating through groups

for name, group in df.groupby('Department'):
    print("Department:", name)
    print(group, "\n")
```

```python
# Merging in Pandas

df1 = pd.DataFrame({
    'Employee': ['A', 'B', 'C'],
    'Department': ['HR', 'IT', 'Sales']
})

df2 = pd.DataFrame({
    'Employee': ['A', 'B', 'C'],
    'Salary': [50000, 60000, 70000]
})

merged = pd.merge(df1, df2, on='Employee')
print(merged)
```

```python
# Types of Merge:
# You can control how the merge is done using the how argument:

# inner	      Default. Only matching keys in both
# outer	      All rows, fill missing with NaN
# left	      All rows from left, matching from right
# right	      All rows from right, matching from left

pd.merge(df1, df2, on='Employee', how='outer')
```

```python
# The how Argument in pd.merge(): The how argument specifies how to handle keys that don’t match between two DataFrames.

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# Second DataFrame
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Score': [85, 90, 95]
})
print(df1)
print('\n')
print(df2)
```

```python
# how='inner' – Only matching rows from both

pd.merge(df1, df2, on='ID', how='inner')

# ➡️ Only IDs 2 and 3 exist in both → included.

```

```python
# how='outer' – All rows from both, NaN where no match

pd.merge(df1, df2, on='ID', how='outer')

# ➡️ Includes all IDs: 1, 2, 3, 4. Missing data filled with NaN.

```

```python
# how='left' – All from df1, matched from df2

pd.merge(df1, df2, on='ID', how='left')

# ➡️ All rows from df1 kept. If df2 doesn't have that ID (like 1), Score is NaN.

```

```python
# how='right' – All from df2, matched from df1

pd.merge(df1, df2, on='ID', how='right')

# ➡️ All rows from df2 are kept. If df1 doesn't have that ID (like 4), Name is NaN.


```

```python
# Convert Categorical Variables

# Methods:
# Label Encoding: Converts categories to numbers (e.g., Male → 0, Female → 1)
# One-Hot Encoding: Creates separate columns for each category

df = pd.DataFrame({
    'Condition': ['Storm', 'Calm', 'Storm', 'Calm']
})

# One-hot encoding
encoded = pd.get_dummies(df, columns=['Condition'])
print(encoded)
```

```python
# pd.get_dummies() output is showing True/False instead of 0/1.
# It's likely because the underlying data type is boolean (bool), 
# which still behaves numerically (True = 1, False = 0) — but is displayed as True/False.

# You can simply convert the dtype using .astype(int):

# One-hot encoding with dtype conversion
df = pd.DataFrame({
    'Condition': ['Storm', 'Calm', 'Storm', 'Calm']
})

# One-hot encoding
encoded = pd.get_dummies(df, columns=['Condition']).astype(int)
print(encoded)
```

```python
# Label Encoding (Ordinal)
# Converts categories into integers (0, 1, 2...).
# Useful when the categories have an order.


df = pd.DataFrame({'Quality': ['Low', 'Medium', 'High', 'Medium']})

# Define order manually
quality_order = {'Low': 0, 'Medium': 1, 'High': 2}
df['Quality_encoded'] = df['Quality'].map(quality_order)
print(df)

# Data has ranking or levels (Low < Medium < High).

```

```python
# One-Hot Encoding (Nominal)
# Creates separate binary columns for each category.
# Use when the categories have no natural order.


df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Red']})

# One-hot encode
df_encoded = pd.get_dummies(df, columns=['Color'])
print(df_encoded)

# Use this when:
# Categories are nominal (e.g., "Red", "Green", "Blue")
# You want to avoid misleading numeric relationships
```

```python
# Binary Encoding / Hashing (Advanced)
# Used for high-cardinality columns (like city names, user IDs).
# Not needed in most beginner/intermediate cases.

df = pd.DataFrame({
    'Weather': ['Sunny', 'Rainy', 'Cloudy', 'Sunny'],
    'Speed': [450, 420, 430, 460]
})

# One-hot encode 'Weather'
df_encoded = pd.get_dummies(df, columns=['Weather'])
print(df_encoded)
```

```python
# Normalize / Scale Data
# Normalize / Scale Data refers to transforming numerical features so that they’re on the same scale

# Common Methods:
# Min-Max Scaling (scales values between 0 and 1)
# Standardization (mean = 0, std = 1)


# Why Normalize or Scale?
# If you have:
    # Feature	      Range
    # Speed	          300 – 800 km/s
    # Density	      0.5 – 10 n/cc

# Then Speed will dominate over Density in distance-based algorithms. Scaling puts them on equal footing.

# Min-Max Scaling (Normalization)

df = pd.DataFrame({
    'Speed': [300, 400, 500, 600],
    'Density': [1.0, 2.5, 4.0, 8.0]
})

# Min-max normalization using pandas
df_scaled = (df - df.min()) / (df.max() - df.min())
print(df_scaled)



# Standardization (Z-score scaling)

# Z-score normalization using pandas
df_standardized = (df - df.mean()) / df.std()
print(df_standardized)

```
